# **Importing required libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Data Preprocessing**

```python
df = pd.read_csv("sample_data/task02ML.csv")
```

```python
df.head()
```

# **Data Analysis**

```python
print(df['Level'].unique())
num_columns = len(df.columns)
print("Total number of columns:", num_columns)
print(df.head(50))
print(df.tail(30))
print(df.info())
print(df.shape)
print(df.size)
print("Number of columns:", df.shape[1])
print("Number of samples:", df.shape[0])
print(df['Level'].value_counts())

# Graphical
df['Level'].value_counts().plot(kind="bar")
plt.show()
subset = df.loc[550:900, 'Gender':'Coughing of Blood']
subset.to_csv("subset1.csv", index=False)

print(subset.describe())  # Mean, Median, Std, etc.
subset2 = df.iloc[-20:, df.columns.get_loc("Shortness of Breath"):df.columns.get_loc("Frequent Cold")+1]
subset2.to_csv("subset2.csv", index=False)

print(subset2.describe())
rows = [110,220,360,440,656,778,202,889]
subset3 = df.loc[rows, :]

avg_alcohol = subset3['Alcohol use'].mean()
print("Average Alcohol use:", avg_alcohol)
```

***TASK 03***

```python
# Age Distribution
df['Age'].hist()
plt.title("Age Distribution")
plt.show()

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    cbar_kws={'shrink': 0.8}
)

plt.title("Correlation Heatmap", fontsize=16, pad=20)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Distribution of Age
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution with KDE")
plt.show()

# Distribution of Frequent cold
sns.histplot(df['Frequent Cold'], kde=True)
plt.title("Frequent Cold Distribution with KDE")
plt.show()

# Outliers in Obesity
sns.boxplot(x=df['Obesity'])
plt.title("Outliers in Obesity")
plt.show()
```
